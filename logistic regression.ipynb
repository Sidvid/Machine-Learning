# breast_cancer by logistic regression

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
pd.DataFrame(cancer.data)
```

```python
cancer.target_names
```

```python
cancer.target.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression().fit(X_train,y_train)
```

```python
print("training score ",logreg.score(X_train,y_train))
print("testing score",logreg.score(X_test,y_test))
```

here we can see that training and testing accuracy is apprx same i.e. our model is going good but also underfit too now we will test on different values of C

```python
logreg100=LogisticRegression(C=100).fit(X_train,y_train)
```

```python
print("training score with c=100 : ",logreg100.score(X_train,y_train))
print("testing score with c=100",logreg100.score(X_test,y_test))
```
